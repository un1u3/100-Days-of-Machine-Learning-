```python
from sklearn.datasets import make_regression
import numpy as np 
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
plt.plot(X,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# Taking slope ass 78.35(from OLS ) and taking intercept = 0 (assume)
y_pred = ((78.35* X) + 0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b = 0 ')
```

```python
m = 78.35
b = 0 
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
# assuming learning rate = 0.1 
lr = 0.1 
step_size = loss_slope * lr
step_size 
```

```python
# calculating rther intercept 
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b = 0 ')
plt.plot(X,y_pred1,color='blue',label='b = 0 ')
```

```python
# Iteration 2 
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope * lr
step_size 
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b = 0 ')
plt.plot(X,y_pred1,color='blue',label='b = 0 ')
plt.plot(X,y_pred2,color='black',label='b = 0 ')
plt.show()

```

```python
# Iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope * lr
step_size 
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='green',label='OLS')
plt.plot(X,y_pred,color='green',label='b = 0 ')
plt.plot(X,y_pred1,color='blue',label='b = 0 ')
plt.plot(X,y_pred2,color='black',label='b = 0 ')
plt.plot(X,y_pred2,color='red',label='b = 0 ')

plt.show()

```

```python
# automating the iterations 
b = 100
m= 78.35
lr = 0.1
epochs = 10 # no of iterations 
for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel()- b)
    b = b - (lr* loss_slope)
    y_pred = m * X + b 
    plt.plot(X,y_pred)
plt.scatter(X,y)
```
